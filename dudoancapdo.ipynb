```python
dfs = []
for name in file_names:
    dfs.append(pd.read_csv(name))
```

```python
from functools import reduce
```

```python
df_final = reduce(lambda left,right: pd.merge(left,right,on=["huyen", "xa", "thon"]), dfs)
```

```python
a = len(df_final.columns[3:])
```

```python
columns = ['huyen', 'xa', 'thon'] + list(range(a))
```

```python
df_final.columns = columns
```

```python
df_final
```

```python
data = df_final[df_final.columns[3:]].to_numpy()
data
```

```python
for i in data:
    print(i,'hello')
```

```python
so_ngay_truoc = 6
d = len(file_names) - so_ngay_truoc
```

```python
x = []
y = []
for i in data:
    for j in range(d):
        x.append(i[j:j+so_ngay_truoc])
        y.append(i[j+so_ngay_truoc])
```

```python
y
```

```python
x_train = []
for x_tr in x:
    x_dem = [0, 0, 0, 0]
    for (i,x_tri) in enumerate(x_tr):
        x_dem[x_tri-1]+=(1+i)
    x_train.append(x_dem) 
```

```python
x_train
```

```python
for (i,x_tri) in enumerate(x_train):
    print(x_tri,' ',y[i])
```

```python
from sklearn.naive_bayes import MultinomialNB
import numpy as np 
```

```python
clf = MultinomialNB()
```

```python
clf.fit(x_train, y)
```
