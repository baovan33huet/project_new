Câu 2: Sử dụng thư viện Numpy, hãy tạo ra ma trận A10x7 với giá trị các phần tử của ma trận A được sinh bởi:

np.arrange()
np.linspace()
np.logspace()

```python
A = np.arange(1, 11).reshape(-1, 1) * np.arange(1, 8)

print(A)
print(A.size)
print(A.ndim)
print(A.shape)

print(A.T)
print(A.T.size)
print(A.T.ndim)
print(A.T.shape)
```

```python
A = np.logspace(1, 2, 10).reshape(-1, 1) * np.logspace(1, 2, 7)

print(A)
print(A.size)
print(A.ndim)
print(A.shape)

print(A.T)
print(A.T.size)
print(A.T.ndim)
print(A.T.shape)
```

```python
def median(X, axis):
    np.sort(X, axis)
    a = len(X) / 2
    if len(X) % 2 == 0:
      medium = ( X[a] + X [a + 1] ) /2
      return medium
    else:
      medium = X[a]
      return medium
```

```python
def count(X):
    max = 0
    dem = 1
    np.sort(X)
    # 1 1 2 3 4 4 4 5
    for i in range(len(X)):
      if X[i] == X[i+1]:
        dem += 1
        if dem > max:
          max = dem
          appear = X[i]
      else:
        dem = 1
    return max, appear
```
